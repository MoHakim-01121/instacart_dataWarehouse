```python
import pandas as pd
```

kita akan menggabungkan kedua dataset terlebih dahulu dan menghapus kolom eval_Set

```python
# load kedua file yang akan kita gabungkan
prior = pd.read_csv("data/order_products__prior.csv")
train = pd.read_csv("data/order_products__train.csv")
```

```python
train.head()
```

```python
# kita gabungkan kedua dataset lalu urutkan berdasarkan order_id
order_products = pd.concat([prior,train])

order_products = order_products.sort_values('order_id')
```

```python
order_products
```

```python
# simpan dalam bentuk csv ke folder /data
order_products.to_csv('data/order_products.csv', index=False)
```

```python
# Opsional: Hapus file lama
import os
os.remove("data/order_products__prior.csv")
os.remove("data/order_products__train.csv")
```

sekarang dataset orders kita akan menggapus kolom eval_set

```python
# load dataset
orders = pd.read_csv("data/orders.csv")
```

```python
# Hapus kolom eval_set (jika ada)
if "eval_set" in orders.columns:
    orders.drop(columns=["eval_set"], inplace=True)
    print("[✓] Kolom 'eval_set' berhasil dihapus.")
else:
    print("[i] Kolom 'eval_set' tidak ditemukan — mungkin sudah dibersihkan sebelumnya.")
```

```python
#cek apakah sudah terhapus kolomnya
orders
```

```python
# Simpan kembali
orders.to_csv("data/orders.csv", index=False)
print("[✓] File orders.csv telah diperbarui tanpa kolom 'eval_set'.")
```
